Data Exploration of the original dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("01_originalDataset_fs180Hz.csv", index_col=0)

```

```python
# Plot random signals of each class
colors = ['g', 'r', 'b', 'm', 'y']

fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(9,9))
for i_cat, cat in enumerate(["N", "S", "V", "F", "Q"]):
      i = np.where(df["class"] == cat)[0]
      color = colors[i_cat]
      if len(i) > 1:
            for  idx in np.random.randint(np.shape(i), size=2):
                  ax1.plot(np.arange(0, 180), df.iloc[i[idx], 6:6+180], '-', color=color, label=df["class"][i[idx]])
                  ax2.plot(np.arange(0, 180), df.iloc[i[idx], 6+180:], '-', color=color, label=df["class"][i[idx]])

ax1.set_title("Random BIH Heartbeats (L2)")
ax1.set_ylabel("ECG-Signal / mV")
ax1.legend()

ax2.set_title("Random BIH Heartbeats (V5)")
ax2.set_xlabel("Time / (samples)")
ax2.set_ylabel("ECG-Signal / mV")
ax2.legend()
```
